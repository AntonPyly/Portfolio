### 2. Знайдіть рекорди кожної статі та кожного підрозділу в кожній із 3 вправ (жим лежачи, присідання та станова тяга)

```python
data_agg.groupby(['sex','division']) \
    .max(['best3squat_kg','best3bench_kg','best3deadlift_kg']) \
    .filter(items=['best3squat_kg','best3bench_kg','best3deadlift_kg'])
```

### 3. Порахуйте кількість перемог кожного учасника, беручи за перемогу 1-е місце. Збережіть результати у вторинному DataFrame.

```python
first_place = data_agg[data_agg.place == '1'].groupby(['division', 'sex', 'name']).agg({'place': 'count'}) \
    .rename(columns={'place': 'number_of_victories'}) 

first_place
```

### 4. Використовуйте DataFrame з вправи 3, щоб показати для кожної комбінації підрозділи та статі, учасника з найбільшою кількістю перемог серед учасників, які будь-коли брали участь у цьому дивізіоні

```python
import itertools

result = pd.DataFrame()

for div, sex in itertools.product(data_agg.division.unique(),data_agg.sex.unique()):
    first_place_name = first_place.loc[(str(div),str(sex)),:].idxmax()[0]
    num = first_place.loc[(str(div),str(sex),str(first_place_name)),:]
    result_buf = pd.DataFrame({'Division': div,
                          'sex': sex,
                          'name': first_place_name,
                          'number_of_wins': num})
    result = pd.concat([result,result_buf])
    
result.reset_index(drop=True, inplace=True)

result
```
