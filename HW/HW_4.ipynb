```python
# начала напишите код для чтения и анализа данных заказа -разделение заказов по строкам и продуктов по «@@@».

with open('orders.txt', 'r') as file:
    data = file.read().split('\n\n')

orders = [order.split('@@@') for order in data]

print(len(orders))
orders[0:2]
                             
```

```python
from collections import defaultdict
```

```python
# Создайте два словаря -один для подсчетакаждого продукта и один для подсчета каждой пары продуктов.
all_product_count = defaultdict(lambda:0)
product_pair = defaultdict(int)

for order in orders:
    for line in order:
        all_product_count[line] += 1
    for i, p1 in enumerate(order[:-1]):
        for p2 in order[i + 1:]:
            product_pair[(p1, p2)] += 1
            product_pair[(p2, p1)] += 1
print(len(all_product_count))
print(len(product_pair))
```

```python
# не вышло сделать пары через combinations. Очень долго идет обработка кода. Можно оптимизировать?
product_pair_2 = defaultdict(int)
from itertools import combinations

comd_order = list(combinations(orders, 2))

for pair in comd_order:
    for line in pair:
        product_pair_2[line] += 1
print(len(product_pair_2))
#
#
```

```python
confidence = {
    (p1, p2): count / all_product_count[p1] for (p1, p2), count in product_pair.items()
}
confidence
```

```python
def associations(min_confidence, min_support, order_list):
    len_order_list = len(order_list)
    for (p1, p2), count in product_pair.items():
        support = count / len_order_list
        conf = confidence[(p1, p2)]
        if conf >= min_confidence and support >= min_support:
            print(f'{p1} => {p2} ({conf*100:.2f}% confidence), {support*100:.2f}% support')
```

```python
associations(.20, 0.01, orders)
```
