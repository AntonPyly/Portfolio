# Matplotlib

Matplotlib — мультиплатформна бібліотека для візуалізації даних, заснована на масивах бібліотеки NumPy та спроектована для роботи з великим стеком SciPy

Деякі функції відображення
* `plt.scatter(x, y, params)` — намалювати точки з координатами з $x$ по горизонтальній осі та з $y$ по вертикальній осі;
* `plt.plot(x, y, params)` — намалювати графік по точках з координатами з $x$ по горизонтальній осі та з $y$ по вертикальній осі. Крапки з'єднаються в тому порядку, в якому вони вказані в цих масивах;
* `plt.fill_between(x, y1, y2, params)` — зафарбувати простір між $y_1$ і $y_2$ за координатами $x$;
* `plt.pcolormesh(x1, x1, y, params)` — зафарбувати простір відповідно до інтенсивності $y$;
* `plt.contour(x1, x1, y, lines)` — намалювати лінії рівня. Потім потрібно застосувати `plt.clabel`.


Вспомогательные функции
* `plt.figure(figsize=(x, y))` — створити графік розміру $(x, y)$;
* `plt.show()` — показати графік;
* `plt.subplot(...)` — додати підграфік;
* `plt.xlim(x_min, x_max)` — встановити межі графіка горизонтальної осі;
* `plt.ylim(y_min, y_max)` — встановити межі графіка вертикальній осі;
* `plt.title(name)` — встановити ім'я графіка;
* `plt.xlabel(name)` — встановити назву горизонтальної осі;
* `plt.ylabel(name)` — встановити назву горизонтальної осі;
* `plt.legend(loc=...)` — зробити легенду у позиції loc;
* `plt.grid()` — додати сітку на графік;
* `plt.savefig(filename)` — зберегти графік у файл.

* plt.plot() → ax.plot()
* plt.legend() → ax.legend()
* plt.xlabel() → ax.set_xlabel()
* plt.ylabel() → ax.set_ylabel()
* plt.xlim() → ax.set_xlim()
* plt.ylim() → ax.set_ylim()
* plt.title() → ax.set_title()

```python
x = np.linspace(0, 10, 1000)

plt.plot(x, x + 4, linestyle='-') # суцільна лінія
plt.plot(x, x + 5, linestyle='--') # штрихова лінія
plt.plot(x, x + 6, linestyle='-.') # штрихпунктирна лінія
plt.plot(x, x + 7, linestyle=':'); # пунктирна лінія

# plt.plot(x, x + 0, linestyle='solid')
# plt.plot(x, x + 1, linestyle='dashed')
# plt.plot(x, x + 2, linestyle='dashdot')
# plt.plot(x, x + 3, linestyle='dotted')
```

```python
x1 = [1,2,3,4,5]
x2 = [i**2 for i in range(5)]

x3 = [1,2,3,4,5]
x4 = [i**3 for i in range(5)]

plt.plot(x1,x2, label = 'pow2', color = 'y', linestyle = '--', marker ='*', linewidth = 3, markersize =15)
plt.plot(x3,x4, label = 'pow3', color = 'black', marker = '.', linewidth = 3, markersize =20)
plt.title('First graph', {'fontname' : 'Arial', 'fontsize' : 25})
plt.xlabel('x-lable')
plt.ylabel('y-lable')
plt.legend()
plt.grid()
plt.show()
```

```python
fig,ax = plt.subplots(figsize = (15,6))
ax.plot(x1,x2, color = 'green')
```
